```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from glob import glob
```

```python
# Loading data sets

mat = pd.read_csv('student-mat.csv', sep=';')
por = pd.read_csv('student-por.csv', sep=';')
student = pd.merge(mat, por, how='outer', suffixes=['_mat','_por'], on=["school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"])
```

```python
# Style
plt.style.use("seaborn-darkgrid")
```

```python
# Reason choosing the school 
# Nominal: 'home': close to home, school: school reputation, 'course': course preference or 'other': other reasons

plt.figure(figsize=(12,8))

school = por['reason'].value_counts()
sns.barplot(school.index, school)

plt.title('Reason choosing the school', fontsize=25)
plt.ylabel('Number of Students', fontsize=15)

plt.savefig('Reason choosing the school.png')
plt.tight_layout()
plt.show()
```

**Question 1:** What is the main reason why students choose their current school? <br>
**Answer:** Most of the students choose their school is because of the course preference that the school offers. 

```python
# Student's Sex & Final Grade
# Binary: 'F' - female or 'M' - male
fig, (ax_mat, ax_por) = plt.subplots(1,2, figsize=(15, 10), sharey=True)

sns.boxplot(x = 'sex', y = 'G3', data=mat, showmeans=True, ax=ax_mat,
           meanprops={"marker":"s","markerfacecolor":"white","markeredgecolor":"black"})
sns.boxplot(x = 'sex', y = 'G3', data=por, showmeans=True, ax=ax_por,
           meanprops={"marker":"s","markerfacecolor":"white","markeredgecolor":"black"})

ax_mat.set_title('Mathematics',fontsize=30)
ax_por.set_title('Portuguese', fontsize=30)

ax_mat.set_ylabel('Final grade', fontsize=18)
ax_mat.set_xlabel("Student's Sex", fontsize=16)
ax_por.set_xlabel("Student's Sex", fontsize=16)

plt.savefig('Students Sex & Final Grade.png')
plt.show()
```

**Question 2:** There is this stereotype that boys are better than girls at S.T.E.M. (Science, technology, engineering, and mathematics), and girls are better than boys at humanity. Is this true? <br>
**Answer:** Based on both data sets, the average and median maths final grades of male students are higher than those of female students. In contrast, the average and median Portuguese final grades of female students are higher than those of male students.

**Question 3:** Do students who want to take higher education have better grades than those who do not? <br>
**Answer:** The average and median final grades in both maths and Portuguese of students who want higher education are higher than those of students who do not want a higher education. This seems to be right. Students who want to go to colleges or universities will need good grades to be accepted; therefore, they will have higher grades than those who don't want to.

**Question 8:** Is there any correlation between the student's absence and the final grade? <br>
**Answer:** Normally, people believe that student attendance (or absence) is a good indicator of student performance. However, based on both data sets (Mathematics and Portuguese), there seems to be no correlation between students' absence and final grade.

```python
# First Period Grade & Final Grade

# Is there any correlation between first period grade and final grade?
# Answer: Yes


fig, (ax_mat, ax_por) = plt.subplots(1,2, figsize=(15, 10), sharey=True)

sns.regplot(x='G1', y='G3', data=mat, ax=ax_mat, color='#e5ae38')
sns.regplot(x='G1', y='G3', data=por, ax=ax_por, color='#008fd5')

ax_mat.set_title('Mathematics',fontsize=30)
ax_por.set_title('Portuguese', fontsize=30)

ax_mat.set_ylabel('Final Grade', fontsize=18)
ax_por.set_ylabel('Final Grade', fontsize=18)
ax_mat.set_xlabel('First Period Grade', fontsize=18)
ax_por.set_xlabel('First Period Grade', fontsize=18)

plt.tight_layout()
plt.savefig('First Period Grade & Final Grade.png')
plt.show()
```

**Question 9:** Is there any correlation between student's first-period grade and final grade? <br>
