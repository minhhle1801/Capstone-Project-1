```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from glob import glob

from scipy.stats import randint

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor

from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

from sklearn.pipeline import Pipeline

from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_validate
from sklearn.model_selection import cross_val_score
```

```python
# Import dataframe

mat = pd.read_csv('student-mat.csv', sep=';')
por = pd.read_csv('student-por.csv', sep=';')
both = pd.merge(mat, por, how='outer', suffixes=['_mat','_por'], on=["school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"])
```

```python
# Encoding dummy variables

math = pd.get_dummies(mat, drop_first=True)
port = pd.get_dummies(por, drop_first=True)
```

# **Mathematics Dataset**

```python
# Train & Test samples for Mathematics

Xm = math.drop('G3', axis=1)
ym = math['G3']

Xm_train, Xm_test, ym_train, ym_test = train_test_split(Xm,ym, test_size=0.2, random_state=24)
```

```python
# Repeated K-Fold cross validator

cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=24)
```

```python
# Linear Regression

linear = LinearRegression()
linear.fit(Xm_train, ym_train)

r2_linear_score = cross_val_score(linear,Xm_train, ym_train,cv=cv, scoring='r2').mean()
print("R2 Cross Validation score: ", r2_linear_score)

mse_linear_score = abs(cross_val_score(linear,Xm_train, ym_train,cv=cv, scoring='neg_mean_squared_error').mean())
print("MSE Cross Validation score: ", mse_linear_score)
```

```python
# Ridge

ridge = Ridge()
ridge.fit(Xm_train,ym_train)

r2_ridge_score = cross_val_score(ridge, Xm_train, ym_train, cv=cv, scoring='r2').mean()
print("R2 Cross Validation Score: ", r2_ridge_score)

mse_ridge_score = abs(cross_val_score(ridge, Xm_train, ym_train, cv=cv, scoring='neg_mean_squared_error').mean())
print("MSE Cross Validation Score: ", mse_ridge_score)
```

```python
# Lasso

lasso = Lasso()
lasso.fit(Xm_train,ym_train)

r2_lasso_score = cross_val_score(lasso, Xm_train, ym_train, cv=cv, scoring='r2').mean()
print("R2 Cross Validation Score: ", r2_lasso_score)

mse_lasso_score = abs(cross_val_score(lasso, Xm_train, ym_train, cv=cv, scoring='neg_mean_squared_error').mean())
print("MSE Cross Validation Score: ", mse_lasso_score)
```

```python
# Support Vector Machines based Regression

steps = [('scaler', StandardScaler()), ('SVR',SVR())]
svr_pipeline = Pipeline(steps)
svr_pipeline.fit(Xm_train, ym_train)

r2_svr_score = cross_val_score(svr_pipeline,Xm_train, ym_train,cv=cv, scoring='r2').mean()
print("R2 Cross Validation Score: ",r2_svr_score)

mse_svr_score = abs(cross_val_score(svr_pipeline,Xm_train, ym_train,cv=cv, scoring='neg_mean_squared_error').mean())
print("MSE Cross Validation Score: ",mse_svr_score)
```

```python
# Decision Tree Regressor

dtr = DecisionTreeRegressor()
dtr.fit(Xm_train, ym_train)

r2_decision_tree_score = cross_val_score(dtr,Xm_train, ym_train,cv=cv, scoring='r2').mean()
print("R2 Cross Validation score: ", r2_decision_tree_score)

mse_decision_tree_score = abs(cross_val_score(dtr,Xm_train, ym_train,cv=cv, scoring='neg_mean_squared_error').mean())
print("MSE Cross Validation score: ", mse_decision_tree_score)
```

```python
# Random Forest Regression

rfr = RandomForestRegressor()
rfr.fit(Xm_train, ym_train)

r2_random_forest_score = cross_val_score(rfr,Xm_train, ym_train,cv=cv, scoring='r2').mean()
print("R2 Cross Validation Score: ", r2_random_forest_score)

mse_random_forest_score = abs(cross_val_score(rfr,Xm_train, ym_train,cv=cv, scoring='neg_mean_squared_error').mean())
print("MSE Cross Validation Score: ", mse_random_forest_score)
```

```python
# Gradient Boosting Regression

gradient = GradientBoostingRegressor()
gradient.fit(Xm_train, ym_train)

r2_gradient_boosting_score = cross_val_score(gradient,Xm_train, ym_train,cv=cv, scoring='r2').mean()
print("R2 Cross Validation Score: ",r2_gradient_boosting_score)

mse_gradient_boosting_score = abs(cross_val_score(gradient,Xm_train, ym_train,cv=cv, scoring='neg_mean_squared_error').mean())
print("MSE Cross Validation Score: ",mse_gradient_boosting_score)
```

```python
# XGBoost

xgboost = XGBRegressor()
xgboost.fit(Xm_train, ym_train)

r2_xgboost_score = cross_val_score(xgboost,Xm_train, ym_train,cv=cv, scoring='r2').mean()
print("R2 Cross Validation Score: ", r2_xgboost_score)

mse_xgboost_score = abs(cross_val_score(xgboost,Xm_train, ym_train,cv=cv, scoring='neg_mean_squared_error').mean())
print("MSE Cross Validation Score: ", mse_xgboost_score)
```

```python
# Lightgbm

light = LGBMRegressor()
light.fit(Xm_train, ym_train)

r2_lightgbm_score = cross_val_score(light,Xm_train, ym_train,cv=cv, scoring='r2').mean()
print("R2 Cross Validation Score: ", r2_lightgbm_score)

mse_lightgbm_score = abs(cross_val_score(light,Xm_train, ym_train,cv=cv, scoring='neg_mean_squared_error').mean())
print("MSE Cross Validation Score: ", mse_lightgbm_score)
```

# **Model Evaluation**

```python
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```

```python
name = ['Linear Regression', 'Ridge', 'Lasso', 'Support Vector Machines', 'Decision Tree',
    'Random Forest','Gradient Boosting', 'XGBoost', 'Lightgbm']

#R2
r2_cv_score = [r2_linear_score, r2_ridge_score, r2_lasso_score, r2_svr_score, r2_decision_tree_score, r2_random_forest_score,
    r2_gradient_boosting_score, r2_xgboost_score, r2_lightgbm_score]

r2_cv_score_round = [round(score, 4) for score in r2_cv_score]

#MSE
mse_cv_score = [mse_linear_score, mse_ridge_score, mse_lasso_score, mse_svr_score, mse_decision_tree_score, mse_random_forest_score,
    mse_gradient_boosting_score, mse_xgboost_score, mse_lightgbm_score]

mse_cv_score_round = [round(score, 4) for score in mse_cv_score]
```

```python
# R2 Cross Validation Score Graph

plt.style.use('seaborn-darkgrid')

fig, ax = plt.subplots(figsize=(8,10))

r2_score_graph = ax.bar(name, r2_cv_score_round)
autolabel(r2_score_graph)

ax.set_ylim(bottom=0.5)    
ax.set_ylabel("Mean of R2 Cross Validation Score", size=12)
ax.set_xlabel('Machine Learning Methods', size=12)
ax.set_title("Mathematics Dataset - Model selection - R2", size = 15)
ax.set_xticks(name)
ax.set_xticklabels(name, rotation=45)

plt.tight_layout()
plt.savefig("Mathematics Dataset - Model selection - R2.png")
plt.show()
```

```python
# MSE Cross Validation Score Graph

fig, ax = plt.subplots(figsize=(8,10))

mse_score_graph = ax.bar(name, mse_cv_score_round)
autolabel(mse_score_graph)

ax.set_ylim(bottom=0.5)    
ax.set_ylabel("Mean of MSE Cross Validation Score", size=12)
ax.set_xlabel('Machine Learning Methods', size=12)
ax.set_title("Mathematics Dataset - Model selection - MSE", size = 15)
ax.set_xticks(name)
ax.set_xticklabels(name, rotation=45)

plt.tight_layout()
plt.savefig("Mathematics Dataset - Model selection - MSE.png")
plt.show()
```
