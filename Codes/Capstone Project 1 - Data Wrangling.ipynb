```python
# Find outlier (Mathematics)

threshold = 3

mat['z_absences'] = np.abs(stats.zscore(mat.absences))
mat['z_G1'] = np.abs(stats.zscore(mat.G1))
mat['z_G2'] = np.abs(stats.zscore(mat.G2))
mat['z_G3'] = np.abs(stats.zscore(mat.G3))
```
